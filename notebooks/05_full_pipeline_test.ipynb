```python
# Imports necess√°rios
import os
import sys
from pathlib import Path

# Configurar path do projeto
PROJECT_PATH = Path("../").resolve()
sys.path.insert(0, str(PROJECT_PATH / "src"))
os.chdir(PROJECT_PATH)

print(f"Project path: {PROJECT_PATH}")
```

## 1. Executar Pipeline via Kedro

```python
from kedro.framework.session import KedroSession
from kedro.framework.startup import bootstrap_project

# Bootstrap do projeto Kedro
bootstrap_project(PROJECT_PATH)

print("Projeto Kedro inicializado!")
```

```python
# Executar pipeline completo
print("Executando pipeline completo...\n")

with KedroSession.create(project_path=PROJECT_PATH) as session:
    result = session.run()

print("\nPipeline executado com sucesso!")
```

## 2. Carregar Resultados

```python
import json

import pandas as pd

# Carregar dados processados
prophet_data = pd.read_parquet(PROJECT_PATH / "data/04_feature/prophet_full.parquet")
predictions = pd.read_parquet(
    PROJECT_PATH / "data/07_model_output/future_predictions.parquet"
)

# Carregar m√©tricas e relat√≥rios
with open(PROJECT_PATH / "data/08_reporting/model_metrics.json") as f:
    metrics = json.load(f)

with open(PROJECT_PATH / "data/08_reporting/forecast_summary.json") as f:
    summary = json.load(f)

print("Dados carregados!")
```

```python
# Exibir m√©tricas do modelo
print("M√©tricas do Modelo:")
print(f"  MAE: ${metrics['mae']:,.2f}")
print(f"  MAPE: {metrics['mape']:.2f}%")
print(f"  RMSE: ${metrics['rmse']:,.2f}")
print(f"  R¬≤: {metrics['r2']:.4f}")
```

```python
# Exibir resumo do forecast
print("\nResumo da Previs√£o:")
print(f"  √öltimo pre√ßo hist√≥rico: ${summary['last_historical_price']:,.2f}")
print(f"  Data: {summary['last_historical_date']}")
print(f"  Dias de previs√£o: {summary['forecast_days']}")

if summary.get("predictions"):
    print("\n  Previs√µes:")
    print(f"    Primeiro dia: ${summary['predictions']['first_day']:,.2f}")
    print(f"    √öltimo dia: ${summary['predictions']['last_day']:,.2f}")
    print(f"    M√°ximo: ${summary['predictions']['max_predicted']:,.2f}")
    print(f"    M√≠nimo: ${summary['predictions']['min_predicted']:,.2f}")
```

## 3. Visualiza√ß√£o Final

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(16, 12))

# Gr√°fico 1: Hist√≥rico completo + previs√µes
ax1 = axes[0]
ax1.plot(prophet_data["ds"], prophet_data["y"], "b-", label="Hist√≥rico", linewidth=1.5)

if not predictions.empty:
    ax1.plot(
        predictions["ds"],
        predictions["predicted_price"],
        "r-",
        label="Previs√£o",
        linewidth=2,
    )
    ax1.fill_between(
        predictions["ds"],
        predictions["predicted_price_lower"],
        predictions["predicted_price_upper"],
        alpha=0.3,
        color="red",
    )

ax1.axvline(x=prophet_data["ds"].max(), color="green", linestyle="--", linewidth=2)
ax1.set_title("Bitcoin - Hist√≥rico Completo e Previs√µes", fontsize=14)
ax1.set_xlabel("Data")
ax1.set_ylabel("Pre√ßo (USD)")
ax1.legend()
ax1.grid(True, alpha=0.3)

# Gr√°fico 2: Zoom nos √∫ltimos 60 dias + previs√µes
ax2 = axes[1]
recent = prophet_data.tail(60)
ax2.plot(recent["ds"], recent["y"], "b-", label="Hist√≥rico (60 dias)", linewidth=2)

if not predictions.empty:
    ax2.plot(
        predictions["ds"],
        predictions["predicted_price"],
        "r-",
        label="Previs√£o",
        linewidth=2,
    )
    ax2.fill_between(
        predictions["ds"],
        predictions["predicted_price_lower"],
        predictions["predicted_price_upper"],
        alpha=0.3,
        color="red",
        label="Intervalo 95%",
    )

ax2.axvline(
    x=prophet_data["ds"].max(), color="green", linestyle="--", label="Hoje", linewidth=2
)
ax2.set_title("Bitcoin - Zoom: √öltimos 60 Dias + Previs√µes", fontsize=14)
ax2.set_xlabel("Data")
ax2.set_ylabel("Pre√ßo (USD)")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Testar API (Opcional)

```python
# Testar a API (certifique-se de que est√° rodando)
import requests

API_URL = "http://localhost:8000"

try:
    # Health check
    response = requests.get(f"{API_URL}/health", timeout=5)
    print("API est√° rodando!")
    print(f"Status: {response.json()}")
except requests.exceptions.ConnectionError:
    print("API n√£o est√° rodando. Execute: python -m crypto_ts_forecast.api.main")
```

```python
# Buscar previs√£o via API
try:
    response = requests.get(f"{API_URL}/api/v1/forecast?days_ahead=7", timeout=30)
    if response.status_code == 200:
        forecast = response.json()
        print("Previs√£o via API (7 dias):")
        for pred in forecast["predictions"]:
            print(f"  {pred['date']}: ${pred['predicted_price']:,.2f}")
    else:
        print(f"Erro: {response.status_code} - {response.text}")
except requests.exceptions.ConnectionError:
    print("API n√£o est√° rodando.")
```

```python
# Buscar pre√ßo atual
try:
    response = requests.get(f"{API_URL}/api/v1/price/current", timeout=10)
    if response.status_code == 200:
        price = response.json()
        print(f"Pre√ßo Atual do Bitcoin: ${price['price']:,.2f}")
        print(f"   Timestamp: {price['timestamp']}")
except requests.exceptions.ConnectionError:
    print("API n√£o est√° rodando.")
```
